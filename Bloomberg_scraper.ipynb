```python
from xbbg import blp
import time

def subscribe_to_stock_updates(ticker):
    """
    Subscribes to real-time updates for a given ticker.
    Prints the received events for a set duration.
    """
    print(f"Subscribing to real-time updates for {ticker}...")

    # Subscribe to market data using the mkdata service
    blp.subscribe(tickers=[ticker], flds=['LAST_PRICE', 'BID', 'ASK'])

    print("Listening for updates for 30 seconds...")
    try:
        # Wait for and print updates
        for _ in range(30):
            time.sleep(1) # Wait for 1 second

            # Use bdp (Bloomberg Data Point) to get the latest values
            data = blp.bdp(tickers=ticker, flds=['LAST_PRICE', 'BID', 'ASK'])
            last_price = data.iloc[0]['last_price']
            bid_price = data.iloc[0]['bid']
            ask_price = data.iloc[0]['ask']

            print(f"[{time.strftime('%H:%M:%S')}] Last Price: {last_price}, Bid: {bid_price}, Ask: {ask_price}")

    except KeyboardInterrupt:
        print("\nSubscription stopped by user.")
    finally:
        # Unsubscribe when done
        blp.unsubscribe(tickers=[ticker])
        print("Unsubscribed from market data.")
# --- Example Usage ---
# Replace with the stock ticker you want to track
# e.g., 'AAPL US Equity', 'GOOGL US Equity'
ticker_to_track = 'GOOGL US Equity'

# Run the subscription
subscribe_to_stock_updates(ticker_to_track)
```

```python
from xbbg import blp
import pandas as pd
from datetime import datetime, timedelta

def get_intraday_tick_history(ticker):
    #Fetches historical intraday tick data for a given ticker for a single day.
    #Bloomberg limits tick data to the most recent 10 days.

    # The bdtick function expects a single date (dt) for which to retrieve tick data.
    # If you need data for a range, you would typically loop through the dates.
    # For this example, we'll fetch data for today's date.
    target_date = datetime.now().date()

    print(f"Fetching intraday tick data for {ticker} on {target_date}...")

    try:
        # blp.bdtick requires `ticker` and `dt` as the first two positional arguments.
        # Optional: start_time and end_time can be added for intraday filtering within the target_date.
        tick_data = blp.bdtick(
            ticker,          # The first required positional argument
            dt=target_date,  # The second required positional argument (can be keyworded for clarity)
            # Example for intraday time filtering (optional):
            # start_time='09:30:00',
            # end_time='16:00:00'
        )

        if not tick_data.empty:
            print("\nSample of fetched tick data:")
            print(tick_data.head())
        else:
            print(f"No tick data could be retrieved for {ticker} on {target_date}. Check ticker and date.")
    except Exception as e:
        print(f"An error occurred while fetching tick data: {e}")
        print("Please ensure you have an active Bloomberg connection and that the ticker is valid.")


# --- Example Usage ---
ticker_for_history = 'SPY US Equity'
get_intraday_tick_history(ticker_for_history)
```
